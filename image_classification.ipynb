```python
import numpy as np
import cv2
import os
import random
import matplotlib.pyplot as plt
```

```python
!unzip /content/drive/MyDrive/leaf.zip
```

```python
DIRECTORY= r"/content/leaf"
CATAGORIES= ['Strawberry_fresh','Strawberry_scrotch']
```

```python
data=[]

for categories in CATAGORIES:
    folder=os.path.join(DIRECTORY,categories)
    label=CATAGORIES.index(categories)


    for img in os.listdir(folder):
        img=os.path.join(folder,img)
        img_arr=cv2.imread(img)
        img_arr=cv2.resize(img_arr,(100,100))

        data.append([img_arr,label])
```

```python
data
```

```python
random.shuffle(data)
```

```python
x=[]
y=[]


for features,label in data:
    x.append(features)
    y.append(label)
```

```python
X= np.array(x)
Y=np.array(y)
```

```python
x
```

```python
X=X/255
```

```python
X
```

```python
X.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense,Activation
```

```python
model=Sequential()
model.add( Conv2D(64,(3,3),input_shape=X.shape[1:],activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add( Conv2D(32,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))



model.add(Flatten())

model.add(Dense(2,activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,Y,epochs=15,validation_split=0.1)
```

```python
model.summary()
```

```python
from keras.preprocessing import image
import numpy as np

img_pred=image.load_img(r"/content/leaf/Strawberry_fresh/3fef560f-5660-4e47-bc07-7a847ceb6a3f___RS_HL 4740.JPG",target_size=(100,100))

img_pred=image.img_to_array(img_pred)
img_pred=np.expand_dims(img_pred, axis=0)


rslt= model.predict(img_pred)

print(rslt)
if rslt[0][0]>rslt[0][1]:
    prediction="Strawberry_fresh"


else:
    prediction="Strawberry_scrotch"
print(prediction)
```
